```python
import sqlite3
import pandas as pd

connection = sqlite3.connect('Daily_transaction_information.db')
df = pd.read_sql('SELECT 日期, 最高價, 最低價, 收盤價 FROM stock', connection, index_col='日期', parse_dates=['日期'])
df.rename(columns = {'日期': 'Date', '最高價': 'High', '最低價': 'Low', '收盤價': 'Price'}, inplace=True)
```
