```python
import requests as req
import pandas as pd
import sqlite3
```

```python
url = 'https://www.twse.com.tw/exchangeReport/STOCK_DAY'
payloads = {
  'response': 'html',
  'date': '20221109',
  'stockNo': '2330'
}
headers = {
  'user-agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_13_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/106.0.5249.119 Safari/537.36'
}
res = req.get(url, params = payloads, headers = headers)
```

```python
df = pd.read_html(res.text)[0]
df.columns = df.columns.droplevel(0)
```

```python
df
```

```python
def convert_date(date):
  year = date.split('/', 1)[0]
  month_day = date.split('/', 1)[1]

  return '{}/{}'.format(int(year) + 1911, month_day)
```

```python
df['日期'] = df['日期'].apply(convert_date)
df
```

```python
df.set_index('日期', inplace=True)
```

```python
df
```

```python
df.to_csv('2330_Daily_transaction_information.csv', encoding='utf-8')
```

```python
df_from_csv = pd.read_csv('2330_Daily_transaction_information.csv', index_col='日期', parse_dates=['日期'])
df_from_csv
```

```python
connection = sqlite3.connect('Daily_transaction_information.db')
df.to_sql('stock', connection, if_exists='replace')
```

```python
df_from_db = pd.read_sql('SELECT * FROM stock', connection, index_col='日期', parse_dates=['日期'])
df_from_db
```
